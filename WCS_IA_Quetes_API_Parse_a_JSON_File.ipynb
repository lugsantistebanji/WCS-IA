# __Parse a JSON file - Challenge__

In this challenge, you will have to import and parse this JSON file, and display a complete DataFrame.

Your DataFrame should have 24 rows.

```python
import json
import pandas as pd
import requests
```

```python
url = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/monthlySalesbyCategoryMultiple.json"

response = requests.get(url)
```

```python
data = json.loads(response.text)
```

```python
data
```

```python
df = pd.json_normalize(response.json()['contents'], record_path='monthlySales', meta=['category', 'region'])
df
```
