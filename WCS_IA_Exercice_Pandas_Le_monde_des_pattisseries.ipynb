3. Affichez les premières lignes pour vérifier.

````python
import pandas as pd

data = {
'Nom': ['Croissant', 'Pain au chocolat', 'Éclair au chocolat',
'Tarte aux fraises', 'Macaron',
'Chausson aux pommes', 'Madeleine', 'Cannelé', 'Baba au rhum', 'Tarte au citron'],
'prix': [1.2, 1.5, 2.0, 2.5, 0.8,
1.8, 0.7, 1.0, 3.0, 2.8],
'ventes': [150, 120, 90, 60, 300,
80, 250, 180, 30, 50],
'Calories': [300, 320, 400, 250, 150,
290, 150, 200, 500, 280],
'Type': ['Viennoiserie', 'Viennoiserie', 'Pâtisserie', 'Pâtisserie', 'Confiserie',
'Viennoiserie', 'Confiserie', 'Confiserie',
'Pâtisserie', 'Pâtisserie'],
'Disponible': ['Oui', 'Oui', 'Oui', 'Non', 'Oui',
'Oui', 'Non', 'Oui', 'Oui', 'Non']
}

pattiseries_df = pd.DataFrame.from_dict(data, orient="columns")
pattiseries_df.set_index("Nom", inplace=True)
pattiseries_df.head(3)
````

### Étape 2 : Calculs et enrichissement

1. **Ajoutez une colonne Revenu journalier (€) :
o Elle est égale à Prix (€) multiplié par Ventes (par jour).**

````python
pattiseries_df = pattiseries_df.assign(
    revenue_journalier=lambda x: x.prix * x.ventes
    )

pattiseries_df
````

2. **Ajoutez une colonne Coût journalier (€) :
o Supposez que le coût de production est de 40 % du prix. Calculez ce coût
comme Prix (€) * 0.4 * Ventes (par jour).**

````python
pattiseries_df = pattiseries_df.assign(
     cout_journalier=lambda x: x.prix*0.4*x.ventes
    )

pattiseries_df
````

3. **Ajoutez une colonne Profit (€) égale à Revenu journalier (€) moins Coût
journalier (€).**

````python
pattiseries_df = pattiseries_df.assign(
    profit=lambda x: x.revenue_journalier - x.cout_journalier)

pattiseries_df
````

### Étape 3 : Analyses avancées

### Étape 4 : Catégorisation et optimisation

1. **Ajoutez une colonne Catégorie :**

- Si les calories sont supérieures à 350, mettez Très calorique.

- Si les calories sont comprises entre 200 et 350, mettez Modéré.

- Sinon, mettez Faible.

````python
def get_category(x):
    category = 'Faible'
    if x > 350:
        category = 'Très calorique'
    elif x > 200:
        category = 'Modéré'
    return category


pattiseries_df['Categorie'] = pattiseries_df['Calories'].apply(get_category)
````
