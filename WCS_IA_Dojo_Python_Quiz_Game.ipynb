##Dojo Python – Quiz Game#

**Instructions :**

- The quiz must contain exactly 3 questions, and therefore 3 answers.
- The player will have 3 tries per question
- If the player answers correctly, his trials are reset for the next question
- Display messages when he loses, wins, when his number of tries changes
- Stop the game if the user loses all his trials
- *Bonus: Set up a point counter for each correct answer and display it*


**Spécifications :**

- Use, at a minimum, a data collection to store expected questions and answers
- Use loops
- Use conditions
- Allow the player to enter their answers
- Stop the game when they have no more tries left (break) or when they have won
- Guide them through the game
- Reduce the number of tries for each wrong answer and reset them after each right
answer.


**Une possible solution**

```python
from typing import List
import random


class Question:

  def __init__(self, question: str, solution: str, other_options: List[str]):
    self._question = question
    self._solution = solution
    self._other_options = other_options

  def get_solution(self) -> str:
    return self._solution

  def get_question(self) -> str:
    return self._question

  def is_good_solution(self, possible_solution: str) -> bool:
    return self._solution == possible_solution

  def random_options(self) -> List[str]:
    random_list = []
    tmp_list = self._other_options + [self._solution]

    while len(random_list) <= len(self._other_options):
      option = random.choice(tmp_list)
      if option not in random_list:
        random_list.append(option)
        if option in tmp_list:
          tmp_list.remove(option)

    return random_list

class Quiz:

  def __init__(self, questions: List[Question], trials: int) -> None:
    self._questions = questions
    self._trials = trials
    self._score = 0


  def run(self):
    trials = 0
    for index, question in enumerate(self._questions):
        options = question.random_options()
        print(f"\n*****{question.get_question()}*****\n")
        print("\n".join(f"{i+1} - {option}" for i, option in enumerate(options))+"\n")

        while trials < self._trials :
            try:
                is_good_answer = False
                answer = int(input("The number of your answer is : "))
                is_good_answer = question.is_good_solution(options[answer-1])
            except:
                print("You need enter a valid number ")
            finally:
                if is_good_answer:
                    print("\n!!!!!Congratulations!!!!!!\n")
                    self._score += 1
                    trials = 0
                    break
                else:
                    print("Wrong answer. Try again.")
                    trials += 1
                    print(f"You have {self._trials - trials}/{self._trials} trials")

        if trials == self._trials:
            break

    if self._score == len(self._questions):
        print("You won")
    else:
        print(f"I'm sorry, you lost. Your final score is {self._score}/{len(self._questions)}")

  def get_score(self) -> int:
    return self._score

  def get_score_message(self) -> str:
    return print(f"You had {self._score} good answers of {len(self._questions)} questions.")


def quiz():
  question1 = Question(
      question="Quel est le plus grand océan du monde ?",
      solution="L'océan Pacifique",
      other_options=["L'océan Atlantique", "L'océan Indien", "L'océan Arctique", "L'océan Antarctique"]
  )

  question2 = Question(
      question="Qui a écrit 'Les Misérables' ?",
      solution="Victor Hugo",
      other_options=["Émile Zola", "Albert Camus", "Gustave Flaubert", "Honoré de Balzac"]
  )

  question3 = Question(
      question="En quelle année l'Homme a-t-il marché pour la première fois sur la Lune ?",
      solution="1969",
      other_options=["1959", "1971", "1982", "1991"]
  )

  new_quiz = Quiz([question1, question2, question3], 3)
  new_quiz.run()
  new_quiz.get_score_message()

quiz()
```
