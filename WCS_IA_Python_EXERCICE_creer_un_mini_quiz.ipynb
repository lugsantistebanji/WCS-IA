Le programme utilisera des variables pour stocker les questions et réponses, des conditions pour vérifier les réponses, des boucles pour permettre plusieurs essais, des opérateurs pour calculer le score, et des fonctions pour organiser le code.

**Une solution type QCM avec POO**

```python
from typing import List
import random


class Question:

  def __init__(self, question: str, solution: str, other_options: List[str]):
    self._question = question
    self._solution = solution
    self._other_options = other_options

  def get_solution(self) -> str:
    return self._solution

  def get_question(self) -> str:
    return self._question

  def is_good_solution(self, possible_solution: str) -> bool:
    return self._solution == possible_solution

  def random_options(self) -> List[str]:
    random_list = []
    tmp_list = self._other_options + [self._solution]

    while len(random_list) <= len(self._other_options):
      option = random.choice(tmp_list)
      if option not in random_list:
        random_list.append(option)
        if option in tmp_list:
          tmp_list.remove(option)

    return random_list

class Quiz:

  def __init__(self, questions: List[Question]) -> None:
    self._questions = questions
    self._score = 0


  def run(self):
    for question in self._questions:
      options = question.random_options()
      print(f"*****{question.get_question()}*****\n")
      print("\n".join(f"{i+1} - {option}" for i, option in enumerate(options)))
      answer = int(input("\nLe numero de votre réponse est: "))

      print();
      if question.is_good_solution(options[answer-1]):
        print("!!!!!Bravo!!!!!!")
        self._score += 1
      else:
        print("Ce n'est pas la bonne réponse")

      print("\n")

  def get_score(self) -> int:
    return self._score

  def get_score_message(self) -> str:
    return print(f"Vous avez eu {self._score} bonne(s) réponse(s) de {len(self._questions)} questions.")


def quiz():
  question1 = Question(
      question="Quel est le plus grand océan du monde ?",
      solution="L'océan Pacifique",
      other_options=["L'océan Atlantique", "L'océan Indien", "L'océan Arctique", "L'océan Antarctique"]
  )

  question2 = Question(
      question="Qui a écrit 'Les Misérables' ?",
      solution="Victor Hugo",
      other_options=["Émile Zola", "Albert Camus", "Gustave Flaubert", "Honoré de Balzac"]
  )

  question3 = Question(
      question="En quelle année l'Homme a-t-il marché pour la première fois sur la Lune ?",
      solution="1969",
      other_options=["1959", "1971", "1982", "1991"]
  )

  new_quiz = Quiz([question1, question2, question3])
  new_quiz.run()
  new_quiz.get_score_message()

quiz()
```
