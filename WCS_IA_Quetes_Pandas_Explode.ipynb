- Which actor is mentioned the most often?


```
import pandas as pd

link = 'https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv'
df = pd.read_csv(link)

df.head(5)
```

```
df.info()
```

```
df.actors_list = df.actors_list.apply(lambda x: eval(x))
df.head(5)
```

```
new_df = df.explode('actors_list').rename(columns={"actors_list": "actor"})
new_df
```

**The actor mentioned the most is:**

```
new_df.groupby('actor')['actor'].count().sort_values(ascending=False).idxmax()
```
