# FizzBuzz avec Compteur

##Contexte :

FizzBuzz est un défi classique de programmation où les nombres de 1 à 100 sont affichés
selon certaines règles. Nous allons ajouter une nouvelle fonctionnalité pour rendre cet
exercice plus intéressant : compter le nombre de fois que "FizzBuzz" apparaît dans la sortie.

##Objectif :

Vous devez écrire un programme en Python qui :

1. Affiche les nombres de 1 à 100 avec les règles suivantes :
    - Si le nombre est un multiple de 3, affichez "Fizz".

    - Si le nombre est un multiple de 5, affichez "Buzz".
    
    - Si le nombre est un multiple de 3 et 5 (donc de 15), affichez "FizzBuzz".
    
    - Sinon, affichez le nombre lui-même.

2. Ajoutez une difficulté :
- Le programme doit également compter combien de fois "FizzBuzz" est affiché.
- À la fin, affichez le nombre total de FizzBuzz".

##Contraintes :

- Vous devez utiliser une boucle pour parcourir
les nombres de 1 à 100.

- Utilisez des conditions (if, elif, else) pour appliquer les règles de FizzBuzz.

- Ajoutez un compteur qui suit le nombre de "FizzBuzz" rencontrés.

##Exemple d'exécution :

**Sortie attendue :**
```python
1
2
Fizz
4
Buzz
Fizz
7
8
Fizz
Buzz
11
Fizz
13
14
FizzBuzz
```

````python
FIZZ = "Fizz"
BUZZ = "Buzz"

def fizz_buzz(number: int) -> str:
    result = ""

    if number % 3 == 0:
        result += FIZZ
    if number % 5 == 0:
        result += BUZZ

    return str(number) if result == "" else result


def main():
    counter = 0

    for number in range(1, 101):
        result = fizz_buzz(number)
        print(result)
        if result == FIZZ + BUZZ:
            counter +=1

    print(f"There are {counter} {FIZZ+BUZZ}")

main()
````
