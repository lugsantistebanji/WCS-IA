# PIVOT TABLE

## It's like in Excel, but with Pandas

There are 2 syntaxes, choose your preferred :
- As a method : YOUR_DATAFRAME.pivot_table()
- As a function : pd.crosstab()

```
# Example
df_weather.pivot_table(values = 'MAX_TEMPERATURE_C', index = 'MONTH', aggfunc = 'mean')
```

---
---

## Create a Pivot table with :
- Months on rows
- the maximum of Max Temperature on values
- Opinion on columns

```
# Hop !
df_weather.pivot_table(values='MAX_TEMPERATURE_C', index='MONTH', columns="OPINION", aggfunc=max)
```

---
---

## Pivot table with :
- Days on rows
- Precipitation on values
- Months on columns


```
# Hop !
df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM', index='DAY', columns="MONTH")
```

---
---

## Pivot table with :
- Months on rows
- the sum of Precipitations on values

**Then display it as a column chart.** (optionnal)

```
# Hop !
pivot_table = df_weather.pivot_table(values='PRECIP_TOTAL_DAY_MM', index='MONTH', aggfunc=sum).rename(columns={'PRECIP_TOTAL_DAY_MM': 'PRECIP_TOTAL_MONTH'})
pivot_table
```

```
pivot_table.plot.bar(rot=0)
```

---
---

## Pivot table with :
- Months on rows
- the average of Max Temperature on a first column of values
- the average of Sunhours on a second column of values

```
# Hop !
pivot_table = df_weather.pivot_table(values=['MAX_TEMPERATURE_C', 'SUNHOUR'], index="MONTH", aggfunc='mean').round(2)

pivot_table
```

---
---

# Chart
Using your favorite library (MatPlotLib, Seaborn, Plotly or what you want), draw a chart from the previous pivot table with :
- Temperature in a blue column chart
- Sunhours in a red line plot

Tip : Once your pivot table created, you can help yourself with the `.reset_index()`  method.

```
pivot_table.reset_index(inplace=True)
pivot_table
```

```
import matplotlib.pyplot as plt


fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('MONTH')
ax1.set_ylabel('MAX_TEMPERATURE_C', color=color)
ax1.bar(x=pivot_table['MONTH'], height=pivot_table['MAX_TEMPERATURE_C'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('SUNHOUR', color=color)
ax2.plot(pivot_table['MONTH'], pivot_table['SUNHOUR'], color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(bottom=2)

fig.tight_layout()
plt.show()
```
