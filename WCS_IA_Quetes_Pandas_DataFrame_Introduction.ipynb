# Mission 2

Create a "fruit_vendors" dataframe that reflects the below example **using a dictionary** data structure

![](https://i.imgur.com/CHPn7ZF.png)

[This resource could help you.](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.from_dict.html)

```
import pandas as pd

dictionary = {
    "2017 Sales":{
        "Apples": 35,
        "Bananas": 21
    },
    "2018 Sales":{
        "Apples": 41,
        "Bananas": 34
    }
}


fruit_vendors = pd.DataFrame.from_dict(dictionary, orient='index')
fruit_vendors
```

# Mission 3

Use a loop to display the values from each column in the dataframe from mission 2.

[This resource could help you](https://www.w3schools.com/python/pandas/ref_df_items.asp)

```
import pandas as pd

dictionary = {
    "2017 Sales":{
        "Apples": 35,
        "Bananas": 21
    },
    "2018 Sales":{
        "Apples": 41,
        "Bananas": 34
    }
}


fruit_vendors = pd.DataFrame.from_dict(dictionary, orient='index')

for col_name, col_values in fruit_vendors.items():
    print(f"Column: {col_name}")
    for value in col_values.values:
        print(value)
    print()
```

# Mission 4

Delete the first row of the dataframe from mission 2

[This resource could help you.](https://sparkbyexamples.com/pandas/pandas-drop-first-row-of-dataframe/#google_vignette)

```
import pandas as pd

dictionary = {
    "2017 Sales":{
        "Apples": 35,
        "Bananas": 21
    },
    "2018 Sales":{
        "Apples": 41,
        "Bananas": 34
    }
}


fruit_vendors = pd.DataFrame.from_dict(dictionary, orient='index')

fruit_vendors.drop(index=fruit_vendors.index[0], inplace=True)
fruit_vendors
```
