## **Mission 1:**

Use apply to create a new column called 'reviews_opinion': if the ratings is between 4.5 and 5, the reviews_opinion is 'very good';

- if the rating is between 4 and 4.5, the reviews_opinion is 'good'; if the rating is between 3 and 4, the reviews_opinions will be 'neutral';

- if the rating is up to 3 the reviews_opinion is going to be 'bad'

```python
books['reviews_opinion'] = books['average_rating'].apply(lambda rating: 'very_good' if rating > 4.5 else 'good' if rating > 4 else 'neutral' if rating > 3 else 'bad')

books[['average_rating','reviews_opinion']]
```

## **Mission 2:**

Using apply with a function, create a column with the number of authors per book

    (note that in the column 'authors', each name of the author is divided with the character '/')

```python
# Add your code here

def count_authors(authors: str) -> int:
    return len(authors.split("/"))

books['nb_authors'] = books['authors'].apply(count_authors)

books[['authors', 'nb_authors']]
```

## **Mission 3:**

Rewrite the same task as Mission 2 using the lamdba function

```python
# Add your code here
books['nb_authors'] = books['authors'].apply(lambda x: len(x.split("/")))
books[['authors', 'nb_authors']]
```

## **Mission 5:**
Obtain the same result of Mission 4 using apply with a function.

    Try both methods to apply to the whole dataset and apply to selected columns

```python
# Add your code here
def get_comment_ration(x):
    if x.ratings_count == 0:
        return 0
    return round(100 * x.text_reviews_count / x.ratings_count, 2)
```

```python

books['comment_ration'] = books.apply(get_comment_ration, axis=1)
books[['text_reviews_count','ratings_count','comment_ration']]
```

```python

books['comment_ration'] = books[['text_reviews_count','ratings_count']].apply(get_comment_ration, axis=1)
books[['text_reviews_count','ratings_count','comment_ration']]
```

# Mission 6:

Obtain the same result of Mission 5 using apply lambda

    Try both methods to apply to the whole dataset and apply to selected columns

```python
# Add your code here
books['comment_ration'] = books.apply(lambda x: round(100 * x.text_reviews_count / x.ratings_count, 2) if x.ratings_count > 0 else 0, axis=1)

books[['text_reviews_count','ratings_count','comment_ration']]
```

```python
books['comment_ration'] = books[['text_reviews_count','ratings_count']].apply(lambda x: round(100 * x.text_reviews_count / x.ratings_count, 2) if x.ratings_count > 0 else 0, axis=1)
books[['text_reviews_count','ratings_count','comment_ration']]
```
