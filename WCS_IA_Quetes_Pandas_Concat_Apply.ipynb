## **Mission 1**

1. Crée une fonction translate qui traduit chacune des 5 valeurs possibles présente dans OPINION en français. Ta fonction doit prendre un str en paramètre, et retourner un autre str avec la traduction.

Par exemple : translate("bad") doit retourner "mauvais". Idem pour very bad, not good not bad, good et very good. Tu peux utiliser un dictionnaire dict, ou bien des conditions imbriquées (le dictionnaire est plus élégant).

    Attention, tu dois prévoir le cas où l'argument fourni n'est pas l'une des 5 valeurs, et retourner alors le texte inchangé. Par exemple : translate("hello") doit retourner "hello".*



**List of phrases of opinion**

```python
opinions = set(list(df2018['OPINION'].unique()) + list(df2019['OPINION'].unique()))
opinions
```

**Translate function**

```python
dict_phares = {
    'bad': 'mauvais',
    'good': 'bien',
    'not good not bad': 'ni bon ni mauvais',
    'unknown': 'inconnu',
    'very bad': 'très mauvais',
    'very good': 'très bien'
}

def translate(phrase: str) -> str:
    if phrase in dict_phares.keys():
        return dict_phares[phrase]
    return phrase
```

## **Mission 2**

1. Applique avec apply ta fonction à la Série OPINION, vérifie que ca fonctionne, puis stocke le résultat dans cette même colonne OPINION.

```python
df2018['OPINION'] = df2018['OPINION'].apply(translate)
df2019['OPINION'] = df2019['OPINION'].apply(translate)
```

2. Utilise apply pour créer une nouvelle colonne MONTH qui contiendra le numéro du mois (de 1 à 12). Tu peux définir une fonction, ou utiliser une fonction lambda.

```python
df2018['MONTH'] = pd.to_datetime(df2018['DATE']).dt.month
df2018['MONTH']
```

## **Mission 3:**

Trace les 4 graphiques suivants pour comparer 2018 et 2019 :
    

_Sur chaque graphique, on pourra lire si 2018 et 2019 sont différents ou similaires, alors utilisez de la couleur pour bien les distinguer !_



```python
import matplotlib.pyplot as plt
```

1. **Histogramme de la vitesse du vent**


```python
fig1, ax1 = plt.subplots()
windspeed2018 = df2018['WINDSPEED_MAX_KMH']
windspeed2019 = df2019['WINDSPEED_MAX_KMH']

colors = ['red', 'blue']
ax1.hist([windspeed2018, windspeed2019], histtype='bar', color=colors, label=[2018, 2019])
ax1.legend(prop={'size': 10})
ax1.set_title('Wind Speed 2018 and 2019 ')
fig1.show()
```

2. **Boxplots pour les précipitations**


```python
precip_2018 = df2018['PRECIP_TOTAL_DAY_MM']
precip_2019 = df2019['PRECIP_TOTAL_DAY_MM']

fig2, ax2 = plt.subplots()

ax2.boxplot([precip_2018, precip_2019], labels=["2018", "2019"])
fig2.show()
```

3. **Scatterplots pour le SUNHOUR et la date**


```python
import matplotlib.dates as mdates

month_day_fmt = mdates.DateFormatter('%b %d')

fig3, ax3 = plt.subplots()
dates2018 = pd.to_datetime(df2018['DATE'])
dates2019 = pd.to_datetime(pd.to_datetime(df2019['DATE']).apply(lambda x: f"2018-{x.month}-{x.day}"))

ax3.scatter(x=dates2018, y=df2018['SUNHOUR'], color="red", label="2018")
ax3.scatter(x=dates2019, y=df2019['SUNHOUR'], color="blue", label="2019")

ax3.xaxis.set_major_formatter(month_day_fmt)

fig3.show()
```

4. **Violinplots pour la température maximale**

```python
max_temp_2018 = df2018['MAX_TEMPERATURE_C']
max_temp_2019 = df2019['MAX_TEMPERATURE_C']

fig4, ax4 = plt.subplots()

ax4.violinplot([max_temp_2018, max_temp_2019], showmedians=True, showextrema=True)
ax4.set_xticks([1,2],labels=["2018", "2019"])
fig4.show()
```
